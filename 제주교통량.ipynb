```python
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
```

```python
df = train_data.copy()
```

```python
except_list = ['id', 'day_of_week', 'road_name', 'start_node_name', 'start_turn_restricted', 'end_node_name', 'end_turn_restricted', 'vehicle_restricted', 'height_restricted']
```

```python
df = df.drop(except_list, axis = 1)
```

```python
y = df['target']
x = df.drop('target', axis = 1)
```

```python
scaler = RobustScaler()
rbs_x = scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1, random_state = 0)
pca_x = pca.fit_transform(rbs_x)
```

```python
pca_x
```

```python
x_train, x_test, y_train, y_test = train_test_split(pca_x, y, test_size = 0.5, random_state=0)
```
