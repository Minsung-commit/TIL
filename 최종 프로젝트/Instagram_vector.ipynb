## ìœ ì‚¬ ì•„ì´í…œ ê²€ìƒ‰

```python
def insta_REC(name, cosine_sim=cosine_sim):
    #ì…ë ¥í•œ ìˆ™ì†Œë¡œë¶€í„° ì¸ë±ìŠ¤ ê°€ì ¸ì˜¤ê¸°
    idx = indices[name]

    # ëª¨ë“  ìˆ™ì†Œì— ëŒ€í•´ì„œ í•´ë‹¹ ìˆ™ì†Œì™€ì˜ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ê¸°
    sim_scores = list(enumerate(cosine_sim[idx]))

    # ìœ ì‚¬ë„ì— ë”°ë¼ ìˆ™ì†Œë“¤ì„ ì •ë ¬
    sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse = True)

    # ê°€ì¥ ìœ ì‚¬í•œ 10ê°œì˜ ìˆ™ì†Œë¥¼ ë°›ì•„ì˜´
    sim_scores = sim_scores[1:11]

    # ê°€ì¥ ìœ ì‚¬í•œ 10ê°œ ìˆ™ì†Œì˜ ì¸ë±ìŠ¤ ë°›ì•„ì˜´
    insta_indices = [i[0] for i in sim_scores]
    
    #ê¸°ì¡´ì— ì½ì–´ë“¤ì¸ ë°ì´í„°ì—ì„œ í•´ë‹¹ ì¸ë±ìŠ¤ì˜ ê°’ë“¤ì„ ê°€ì ¸ì˜¨ë‹¤. ê·¸ë¦¬ê³  ìŠ¤ì½”ì–´ ì—´ì„ ì¶”ê°€í•˜ì—¬ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë„ í™•ì¸í•  ìˆ˜ ìˆê²Œ í•œë‹¤.
    result_data = data.iloc[insta_indices].copy()
    result_data['score'] = [i[1] for i in sim_scores]
    
    # ì½ì–´ë“¤ì¸ ë°ì´í„°ì—ì„œ ì¤„ê±°ë¦¬ ë¶€ë¶„ë§Œ ì œê±°, ì œëª©ê³¼ ìŠ¤ì½”ì–´ë§Œ ë³´ì´ê²Œ í•¨
    del result_data[['content','token_nolist','token_content']]

    # ê°€ì¥ ìœ ì‚¬í•œ 10ê°œì˜ ìˆ™ì†Œì˜ ì œëª©ì„ ë¦¬í„´
    return result_data
```

```python
def get_sentence_mean_vector(morphs):
    vector = []
    for i in morphs:
        try:
            vector.append(embedding_model.wv[i])
        except KeyError as e:
            pass
    try:
        return np.mean(vector, axis=0)
    except IndexError as e:
        pass
```
