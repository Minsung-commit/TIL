```python
# 모듈 호출
import pandas as pd
import numpy as np
from impyute.imputation.cs import fast_knn
from fancyimpute import KNN
```

```python
#데이터 불러오기
a = pd.read_csv('./total_08_new.csv')
sample = a.loc[a.new_emp == '자영업자']
```

```python
sample.replace({'0' : np.NaN}, inplace = True) # 0 값 변환
sample
```

```python
df = pd.get_dummies(sample.iloc[:, 13], prefix=None, prefix_sep='_', dummy_na=True, columns=["n","y","null"], sparse=False, drop_first=False, dtype='float') # 1개 변수 원 핫 인코딩
```

```python
df.columns = ["n","y","null"] #컬럼명 설정
```

```python
df
```

```python
df.loc[df.null == 1, ["n", "y"]] = np.NaN #NaN 값 입력
```

```python
df = df.drop('null', axis = 1) #null 칼럼 제거
```

```python
df
```

```python
np_imputed=fast_knn(df.values, k=5)# KNN 학습 
df_imputed = pd.DataFrame(np_imputed)
```

```python
knn_imputer = KNN()
df_knn = df.copy()
df_knn.iloc[:,:] = knn_imputer.fit_transform(df_knn)
```
