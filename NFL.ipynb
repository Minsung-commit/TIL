```python
# 범주별 구성비율에 따른 결측치 대체
def type_fill_na():
  rnd_num = np.random.uniform(0,1,1) #균등분포로부터 난수 생성

  if rnd_num > 0.93:
    x = 'retractable'
  elif rnd_num > 0.77:
    x = 'indoor'
  else :
    x = 'outdoor'

  return x
```

```python
def weather_fill_na():
  rnd_num = np.random.uniform(0,1,1) #균등분포로부터 난수 생성

  if rnd_num > 0.82:
    x = 'warm'
  elif rnd_num > 0.61:
    x = 'moderate'
  elif rnd_num > 0.38:
    x = 'dome'
  else :
    x = 'cold'

  return x
```

```python
## 데이터 유형이 str이라 isnull를 통한 값 위치 특정이 어려워, 우선 null값을 ""으로 대체한 후 for을 통해 값 대체 진행
t_stadiums.stadium_type.fillna("", inplace = True)
t_stadiums.stadium_weather_type.fillna("", inplace = True)
```

```python
for i in range(len(t_stadiums.stadium_type)): # 결측값 대치 
  if t_stadiums.stadium_type.iloc[i] == "":
    t_stadiums.stadium_type.iloc[i] = type_fill_na()
```

```python
for i in range(len(t_stadiums.stadium_weather_type)):
  if t_stadiums.stadium_weather_type.iloc[i] == "":
    t_stadiums.stadium_weather_type.iloc[i] = weather_fill_na()
```

```python
df1 = t_stadiums
df2 = score[['score_home', 'score_away','stadium','weather_temperature', 'weather_wind_mph']]
```

```python
total = pd.merge(df1, df2, how='inner', left_on='stadium_name', right_on = 'stadium')
```

```python
total.drop('stadium', axis = 1, inplace = True)
```

```python
total
```

```python
total['win'] = np.NaN
```

```python
for i in range(len(total)):
  if total.score_home.iloc[i] > total.score_away.iloc[i]:
    total.win.iloc[i] = 1
  elif total.score_home.iloc[i] < total.score_away.iloc[i]:
    total.win.iloc[i] = 0
```

```python
total.win.fillna(1, inplace = True)
```

```python
# total.drop('weather_humidity', axis = 1, inplace = True)
total.dropna(axis = 0, subset = ['score_home'], inplace = True)
total.dropna(axis = 0, subset = ['weather_wind_mph'], inplace = True)
```

```python
total.isnull().sum() ## 확인 
```

```python
total.corr()
```

```python
s_type = pd.get_dummies(total.stadium_type, drop_first=True)
w_type = pd.get_dummies(total.stadium_weather_type, drop_first=True)
location = pd.get_dummies(total.stadium_location, drop_first=True)
```

```python
sub_total = pd.concat([s_type, w_type, location], axis = 1)
```

```python
ss_total = total[['weather_temperature', 'weather_wind_mph', 'win']]
```

```python
df = pd.concat([ss_total, sub_total], axis = 1)
```

```python
df
```

```python
y = df.win
x = df.drop('win', axis = 1)
```

```python
# x = pd.DataFrame(x)
```

```python
x
```

```python
np.set_printoptions(precision=6, suppress=True)
pd.options.display.float_format = '{:.5f}'.format
pd.reset_option('display.float_format')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # 주성분을 몇개로 할지 결정
x_pca = pca.fit_transform(x)
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
nb = GaussianNB()
knn = KNeighborsClassifier(n_neighbors=5)
svm = SVC()
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(n_estimators=5)
```

```python
x_pca.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size = 0.3, stratify=y) 
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
knn.fit(x_train, y_train)
nb.fit(x_train, y_train)
svm.fit(x_train, y_train)
lr.fit(x_train, y_train)
dt.fit(x_train, y_train)
rf.fit(x_train, y_train)
```

```python
knn_prd = knn.predict(x_test)
nb_prd = nb.predict(x_test)
svm_prd = svm.predict(x_test)
lr_prd = lr.predict(x_test)
dt_prd = dt.predict(x_test)
rf_prd = rf.predict(x_test)
```

```python
print("=== KNN Classification Report ===")
print(classification_report(y_test, knn_prd))
print('\n')
```

```python

print("=== KNN Classification Report ===")
print(classification_report(y_test, knn_prd))
print('\n')
print("=== NB Classification Report ===")
print(classification_report(y_test, nb_prd))
print('\n')
print("=== SVM Classification Report ===")
print(classification_report(y_test, svm_prd))
print('\n')
print("=== LR Classification Report ===")
print(classification_report(y_test, lr_prd))
print('\n')
print("=== DT Classification Report ===")
print(classification_report(y_test, dt_prd))
print('\n')
print("=== RF Classification Report ===")
print(classification_report(y_test, rf_prd))
print('\n')
```

```python
knn_score = accuracy_score(y_test, knn_prd)
nb_score = accuracy_score(y_test, nb_prd)
svm_score = accuracy_score(y_test, svm_prd)
lr_score = accuracy_score(y_test, lr_prd)
dt_score = accuracy_score(y_test, dt_prd)
rf_score = accuracy_score(y_test, rf_prd)
```
