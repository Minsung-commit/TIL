```python
rest_behappyhere_pd = rest_behappyhere_DFP.toPandas()
rest_behappyhere_pd
```

```python
import re

def extract(x):
    p = re.compile('\"[가-힣]+\"')
    m = p.findall(x)
    return m[0]

rest_behappyhere_pd.content.apply(lambda x : extract(x))
```
