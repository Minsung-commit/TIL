```python
import pandas as pd
```

```python
import numpy as np
```

```python
Seoul_df = pd.read_csv('./Seoul_Company_num.csv', encoding='cp949')
```

```python
Seoul_df.head(20)
```

```python
Seoul_df.columns = ('Field', 'Size', 2017, 2018, 2019 )
```

```python
size = Seoul_df.Size.unique()
size_list = size.tolist()
```

```python
# Seoul_df
```

```python
df1 = Seoul_df[Seoul_df.Field == '전체']
```

```python
df2 = Seoul_df[Seoul_df.Field == 'G.도매 및 소매업(45~47)']
```

```python
df3 = Seoul_df[Seoul_df.Field == 'H.운수 및 창고업(49~52)']
```

```python
df4 = Seoul_df[Seoul_df.Field == 'I.숙박 및 음식점업(55~56)']
```

```python
df5, df6, df7, df8 = Seoul_df[165:176], Seoul_df[176:187], Seoul_df[187:198], Seoul_df[198:209]
```

```python
df5, df6, df7, df8 = df5.iloc[:, 2:], df6.iloc[:, 2:],df7.iloc[:, 2:],df8.iloc[:, 2:]
```

```python
df5.reset_index(drop = True,inplace = True)
```

```python
df6.reset_index(drop = True,inplace = True)
```

```python
df7.reset_index(drop = True,inplace = True)
```

```python
df8.reset_index(drop = True,inplace = True)
```

```python
df_services = df5 + df6 + df7 + df8
```

```python
df_services 
```

```python
df1 = df1.drop(['Field','Size'],axis=1).reset_index(drop = True)
```

```python
df2 = df2.drop(['Field','Size'],axis=1).reset_index(drop = True)
```

```python
df3 = df3.drop(['Field','Size'],axis=1).reset_index(drop = True)
```

```python
df4 = df4.drop(['Field','Size'],axis=1).reset_index(drop = True)
```

```python
df_all = df1.transpose()
df_sales = df2.transpose()
df_trans = df3.transpose()
df_accom = df4.transpose()
df_services = df_services.transpose()
```

```python
df_sales
```

```python
result = pd.concat([df_all, df_sales, df_trans, df_accom, df_services])
```

```python
result.columns = size_list
```

```python
result = result.reset_index().rename(columns={"index": "year"})
```

```python
field = ['all','all','all', 'whole_retail','whole_retail','whole_retail', 'transport','transport','transport', 'accommodation','accommodation','accommodation', 'services','services','services']
```

```python
col1 = result.columns.tolist()
```

```python
field = pd.DataFrame(field, columns = ['field'])
# field = field.rename(columns={0:'field'})
```

```python
result = pd.concat([result,field],axis=1)
```

```python
col1 = result.columns[:-1].to_list()
col2 = result.columns[-1:].to_list()
new_col = col2 + col1
result = result[new_col]
```

```python
result.to_csv('.\\company_num.csv', sep=',', na_rep='NaN', index = False)
```
